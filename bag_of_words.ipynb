```python
def review_to_words(raw_review):
    review_text=BeautifulSoup(raw_review).get_text() # convert HTML
    words_only=re.sub("[^a-zA-Z]"," ", review_text)    #remove number and punctuatuions
    lower_words=words_only.lower()        #convert to lower cases
    words=lower_words.split()      #splitting into individual words
    stops=set(stopwords.words("english"))        #converting words from stoplist to a set
    words=[x for x in words if not x in stops]      #removing stopwords from words list
    string_review=" ".join(words)   #Joining words to make a string
    return string_review
```

```python
print review_to_words(train['review'][0])
```

```python
num_reviews=train["review"].size
print num_reviews

clean_reviews=[]

for i in xrange(num_reviews):
    if(i%5000==0):
        print "At ",i,"th review"
    
    review=review_to_words(train["review"][i])
    
    clean_reviews.append(review)

print len(clean_reviews)
print "Done"
```

```python
#Creating bag of words

from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer(analyzer="word", stop_words=None, tokenizer=None, preprocessor=None, max_features=5000)
train_data_features=vectorizer.fit_transform(clean_reviews)

train_data_features=train_data_features.toarray()
```

```python
print train_data_features.shape
```

```python
vocab=vectorizer.get_feature_names()
print vocab
```

```python
import numpy as np

#sum up counts
dist=np.sum(train_data_features,axis=0)

for feature,count in zip(train_data_features,dist):
    print feature, count
```

```python
#Training the random forest

from sklearn.ensemble import RandomForestClassifier

forest=RandomForestClassifier(n_estimators=100)

forest=forest.fit(train_data_features, train["sentiment"])
```

```python
#import testing data

test=pd.read_csv("testData.tsv",delimiter="\t", header=0, quoting=3)

print test.shape

clean_test_review=[]

#cleaning data

for i in xrange(0,num_reviews):
    clean_review=review_to_words(test["review"][i])
    clean_test_review.append(clean_review)
    if(i%5000==0):
        print "At ",i,"th review"

test_data_features=vectorizer.transform(clean_test_review)
test_data_features=test_data_features.toarray()

result=forest.predict(test_data_features)

output=pd.DataFrame(data={"id":test["id"], "sentiment":result})

output.to_csv("Bag_of_Words_model.csv",index=False,quoting=3)
```
